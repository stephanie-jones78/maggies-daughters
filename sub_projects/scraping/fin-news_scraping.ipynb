# Some Introductory Code to Automate Financial News
Using [CNBC API](https://rapidapi.com/apidojo/api/cnbc/)

```python
import pandas as pd
import requests
import env
```

```python
news_url = 'https://cnbc.p.rapidapi.com/news/v2/list'

news_query_str = {"franchiseId":"19794221","count":"25"}

headers = {'X-RapidAPI-Key': env.cnbc_key, 'X-RapidAPI-Host': env.cnbc_host}
```

```python
response = requests.request('GET', news_url, headers = headers, params = news_query_str)

print(response.text)
```
