```python
def daily_ticker_data(tick_symbol):
    # getting api url
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&datatype=csv&apikey={env.key}'
    
    # requesting and decoding data
    with requests.Session() as s:
        s.post(url, data = requests.get(url))
        download = s.get(url)
        decoded = download.content.decode('utf-8')
    
    # reading requested data to pandas df
    df = pd.read_csv(StringIO(decoded))
    
    # adding column for ticker
    df['ticker'] = tick_symbol
    
    return df
```

```python
daily_ticker_data('DIS')
```
